Adding new columns
Create happiness categories

```python
# 1. Create a new column called Happiness Category that categorizes countries into
# three categories based on their Score:

# Low − (Score < 4)
# Medium − (4 ≤ Score ≤ 6)
# High − (Score > 6)

# Create a new column for Happiness Category
def  happiness_category(score):
    if score >= 6:
        return 'High'
    elif score >= 5:
        return 'Medium'
    else:
        return 'Low'

data['Happiness Category'] = data['score'].apply(happiness_category)
data.head()
```
